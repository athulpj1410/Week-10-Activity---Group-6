# Check the outliers and handle outliers in performance score using Percentiles

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure()
plt.boxplot(df['performance_score'])
plt.title("Performance Score",fontsize = 16)
plt.show()
```

## Since there is one outlier in the performance score feature we need to remove it

```python
import numpy as np

q1 = np.percentile(df['performance_score'],25, interpolation = "midpoint")

q2 = np.percentile(df['performance_score'],50, interpolation = "midpoint")

q3 = np.percentile(df['performance_score'],75, interpolation = "midpoint")

print(" Q1 = {0} \n Q2 = {1} \n Q3 = {2}".format(q1,q2,q3))
```

```python
df['performance_score'].median()
```

```python
iqr = q3 - q1
print("\n",iqr,"\n\n\n")
lowerLimit = q1 - (1.5*iqr)
upperLimit = q3 + (1.5*iqr)
print(" Lower Limit = {0} \n Upper Limit = {1}".format(lowerLimit,upperLimit))
```

```python
outlier = []

for i in df['performance_score']:
    if (i>upperLimit or i<lowerLimit):
        outlier.append(i)
        
print(outlier)
```

```python
index1 = df['performance_score']<lowerLimit
df.loc[index1].index
```
